```python

plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
sns.histplot(df['sepal_length'])
plt.subplot(2,2,2)
sns.kdeplot(df['sepal_length'])
plt.subplot(2,2,3)
sns.boxplot(df['sepal_length'])
plt.show()
```

```python
def printfunc(name):
  plt.figure(figsize=(12,7))
  plt.subplot(2,2,1)
  sns.histplot(df[name])
  plt.subplot(2,2,2)
  sns.kdeplot(df[name])
  plt.subplot(2,2,3)
  sns.boxplot(df[name])
  plt.show()
```

```python
printfunc("sepal_width")
```

```python
for feature in df.columns:
  if(feature == 'species'):
    continue
  print(feature)
  
  printfunc(feature)
```
