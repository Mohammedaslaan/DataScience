```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
gr = iris.groupby('species')
```

```python
gr.groups
```

```python
gr_setosa = gr.get_group('setosa')
```

```python
gr_setosa.describe()
```

```python
gr_versicolor = gr.get_group('versicolor')
gr_versicolor.describe()
```

```python
gr_verginica = gr.get_group('virginica')
gr_verginica.describe()
```
