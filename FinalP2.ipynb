```python
df['math score'].quantile(0.25)
IQR = df['math score'].quantile(0.75) - df['math score'].quantile(0.25)
```

```python
min_math = df['math score'].quantile(0.25) - 1.5 * IQR
```

```python
df = df[(df['math score']>= min_math)]
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
df.boxplot('math score')
plt.subplot(2,2,2)
sns.histplot(df['math score'])
plt.subplot(2,2,3)
sns.kdeplot(df['math score'])
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
df.boxplot('reading score')
plt.subplot(2,2,2)
sns.histplot(df['reading score'])
plt.subplot(2,2,3)
sns.kdeplot(df['reading score'])
```

```python
IQR = df['reading score'].quantile(.75) - df['reading score'].quantile(0.25)
min = df['reading score'].quantile(.25) - 1.5* IQR 
min
```

```python
df = df[(df['reading score']>= min)]
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
df.boxplot('reading score')
plt.subplot(2,2,2)
sns.histplot(df['reading score'])
plt.subplot(2,2,3)
sns.kdeplot(df['reading score'])
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
df.boxplot('writing score')
plt.subplot(2,2,2)
sns.histplot(df['writing score'])
plt.subplot(2,2,3)
sns.kdeplot(df['writing score'])
```
